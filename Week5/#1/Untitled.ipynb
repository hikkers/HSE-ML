```python
import pandas as pd
import numpy as np

data = pd.read_csv('abalone.csv')
```

```python
data.head()
```

```python
data['Sex'] = data['Sex'].map(lambda x: 1 if x == 'M' else(-1 if x == 'F' else 0))
data
```

```python
X = data.drop(columns=['Rings'])
y = data['Rings']
```

```python
X
```

```python
y
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score

kf = KFold(n_splits=5, shuffle=True, random_state=1)
#res = {'n_trees':, 'r2_score':, 'cv_score':}
#res = {'n_trees':[], 'r2_score':[], 'cv_score':[]}
res = {'n_trees':[], 'cv_score':[]}

for i in range(1, 51):
    rf = RandomForestRegressor(n_estimators=i)
    cv = cross_val_score(rf, X, y, cv=kf, scoring='r2').mean()
    rf.fit(X, y)
    pred = rf.predict(X)
#    r2 = r2_score(y, pred)
    res['n_trees'].append(i)
#    res['r2_score'].append(r2)
    res['cv_score'].append(cv)


    
```

```python
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df = df.sort_values(by=['cv_score'], axis=0, ascending=False)
df
```
