```python
df6 = data_set.loc[data_set['Sex'] == 'female']
df6
```

```python
df6 = df6[['PassengerId', 'Name']]
df6
```

```python
import re
#\w+, (Miss\. |Mrs\. |Mr\. ) select all letters before comma and select title from the group
#\w+, select all letters before comma
#\(([^)]+)\) select all between rounds

def clean_name(name):
    p = re.search("^[^,]+, (.*)", name)

    if p:
        name = p.group(1)

    p = re.search("\(([^)]+)\)", name)

    if p:
        name = p.group(1)


    name = re.sub("(Miss\. |Mrs\. |Mr\. )", "", name)
    name = name.split(' ')[0].replace('"', '')
    return name
```

```python
names = df6['Name'].apply(clean_name)
names
```
