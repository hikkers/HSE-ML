```python
import pandas as pd
import numpy as np

train = pd.read_csv('close_prices.csv', index_col='date')
train.head()
```

```python
from datetime import datetime

train['date'] = train['date'].apply(lambda x: x.replace('-',''))
train
```

```python
from sklearn.decomposition import PCA

pca_mdl = PCA(n_components=10)
pca_mdl.fit(train)
```

```python
pca_mdl.explained_variance_ratio_[0:4].sum()
```

```python
pca_mdl.transform(train)
```

```python
comp = pd.DataFrame(pca_mdl.transform(train)[:,0])
comp
```

```python
test = pd.read_csv('djia_index.csv', index_col='date')
test
```

```python
coef = np.corrcoef(comp.T, test.T)[1,0]
print("Corrcoef %0.2f" % coef)
```

```python
print(train.columns[ np.argmax( pca_mdl.components_[1])])
```
